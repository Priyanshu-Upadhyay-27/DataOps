Relational Plots are used to do multi variate analysis

Scatter Plot is used mainly to depict relationship between numerical data.
There are 2 ways to draw a scatterplot in seaborn

1. relplot [Figure Level Function]
2. scatterplor[Axes Level Function]

As we will see, these visualizations can be quite helpful because they use simple and easily-
understood representations of data even on complex data by mapping up to three additional
variables using the semantics of hue, size, and style.

```python
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter")
#scatter is default value of kind, other is "line". 
```

```python
# hue parameter = It adds an information of smoker or particular column by changing the color of dots.(categorical data)
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips)
```

```python
# style parameter = It also adds an information of sex or particular column by changing the style or shape of dots.(Categorical Data)
sns.relplot(x="total_bill", y="tip",hue="smoker", style="sex", data=tips)
```

```python
# size parameter = It also adds an information about size or particular column by changing the size of dots or marks(numerical data)
sns.relplot(x="total_bill", y='tip', hue="smoker", style="sex", size="size", data=tips, sizes=(15,100))
# you can also use "sizes" parameter which inputs tuple having two values indicating a range for new size
```

As you can see, with the help of seaborn, we can get large amount of information from a single plot.

```python
# scatter function
sns.scatterplot(x="total_bill", y='tip', hue="smoker", style="sex", size="size", data=tips, sizes=(15,100))
```

# Categorical Scatterplot (catplot) 

```python
# use categorical data on x-axis
# use numerical data on y-axis to plot categorical plot
plt.style.use("fivethirtyeight")
```

## Strip Plot

Strip Plot is used to draw a plot which compares two or more categorical data with numerical data, and plot it using dots.

```python
sns.catplot(x="day", y="tip", data=tips, kind="strip")
# It plots a stripplot by default as below
# Categorical vs Numerical Data Plot
```

```python
sns.catplot(x="day", y="tip", data=tips, hue="time", kind="strip",jitter=0.21)
# jitter paramter is used to make some noise or space between the dots, by default it has some value.
```

## Swarm Plot

It is similar to strip plot but it spreads the dots if they overlap a little bit too. 

```python
sns.catplot(x="day", y="tip", kind="swarm", data=tips, size=4)
#size is used here to decerease the size of dots since all dots ar unable to print.
```

```python
sns.swarmplot( x="day", y="tip", hue="sex", data=tips)
#size and style parameter are not available for both swarm and strip plot and the size parameter used above is for size of all the dots.
```

## Boxplot

It is a standardized way of displaying the data in the form of 5 main types, which are minimum value, maximum value, quartile range 1, quartile range 2 and the median. Some other things are also there like outliers, whiskers etc.
It helps in displaying about the data that is it symmetrical, tightly grouped or how it changes etc.

```python
#Univariate Analysis : Only one axis is defined for numerical data only.
sns.boxplot(y=tips["tip"])
# column of tips dataframe is used as argument
```

```python
sns.boxplot(x="total_bill", data=tips)
```

```python
# Multivariate Analysis : we define both axis, one as categorical and other as numerical
sns.catplot(x="day", y="total_bill", kind="box", data=tips)
```

```python
sns.boxplot(x="time", y="tip", data=tips)
```

```python
sns.catplot(x="day", y="total_bill", data=tips, hue="sex", kind="box")
```

## Violin Plot

These plots are used to plot numerical data and can be considered as a combination of box plot and kernel density plot.
Advantage of this plot is that, besides providing the 5 number summary, it also shows the entire distribution of data.
They can be used to compare the distribution of data of a variable across some categories.

```python
sns.violinplot(x=tips["total_bill"])
```

```python
#new=["priyavrat", "mohu", "mangla", "gurjar"]
sns.catplot(x="day", y="total_bill", data=tips, kind="violin" )
plt.grid(True)
# From the area distribution of violin plots, we can infer the density of total bills paid on that day.
```

```python
# Create a kernel density plot for the overall 'total_bill' column
sns.kdeplot(data=tips, x="total_bill", shade=True, label="Overall")

# Create a kernel density plot for Thursdays only
thursday_tips = tips[tips["day"] == "Thur"]
sns.kdeplot(data=thursday_tips, x="total_bill", shade=True, label="Thursday")

# Add a title and axis labels
plt.title("Kernel Density Estimate of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Density")

# Add a legend
plt.legend()
# Above voilin plot can be verified from this kde, that on thursday more number of people had eaten in the restaurant.
```

```python
sns.catplot(data=tips, x="day", y="total_bill", hue="smoker", kind="violin", split=True)
#Split is used to combine the half part of both type of data, since data distribution remains unchanged as it is symmetrical to both sides in violin plots.
```

## Barplots

It is used to infer central tendency from a data, in which one axis is categorical and other is numerical.
A bar plot is a graphical representation of data that uses rectangular bars to show the frequencies or counts of different categories. The height of each bar represents the value of the category, and the width of each bar is typically the same. Bar plots can be used to compare different categories of data, or to show how a category has changed over time.

```python
sns.catplot(x="sex", y="total_bill", data=tips, kind="bar")
#Bar heights give you avg total bill of particular category(male or female).
```

```python
sns.catplot(x="sex", y="total_bill", kind="bar", data=tips, hue="time")
```

```python
sns.catplot(x="sex", y="total_bill", kind="bar", data=tips,  hue="time",estimator=np.median)
# estimator is used to change the operation like from mean to median or variance or standard deviation etc.
```

## Count Plot

They are used to count the number of observations, they work like histogram but on categorical data.

```python
sns.catplot(x="sex", kind="count", data=tips)
```

```python
sns.catplot(x="smoker", kind="count", data=tips, hue="size")
# Group of 2 people are highest in number of smoking and non-smoking category.
```

# Heatmaps

It is a graphical representation of data where the individual values contained in a matrix are represented as colors. 

```python
flights = sns.load_dataset("flights")
```

```python
flights.head(24)
```

```python
x=flights.pivot_table(index="year", columns="month", values="passengers", aggfunc='sum')
x
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x, cbar=False, linewidth=0.4)
```

```python
sns.heatmap(x, cbar=True, linewidth=1, annot=True, fmt="d", cmap="summer")
#Here annot is used to print exact values of each blocks and d represents integers to be printed.
#Here cmap is used to change the tone of color.
```

# Cluster Map

It is a heatmap but here similar rows and similar columns are grouped together side by side.Lines used by side and top of clustermaps are dendrograms which represents hiearchay or grouping.

```python
sns.clustermap(x)
```

```python
sns.clustermap(x, z_score=0, metric="correlation",annot=True)
# metric parameter uses different types of distances(default:eucledean), according to which rows and columns are clustered differently.
# z_score refers to standardized score that measures how many standard deviations a data point is from the mean of the data set.
```

```python
sns.clustermap(x, z_score=0, metric="correlation",annot=True, row_cluster=False)
#row_cluster= false, removes the clustering on the basis of rows.
```

# Joint Plots

Jointplot creates a multi-panel figure i.e. combining two graphs that include both the bivariate (or joint) relationship between two variables along with the univariate (or marginal) distribution of each on separate axes.

```python
sns.jointplot(x="total_bill" , y="tip", data=tips)#scatterplot(default)
plt.rcdefaults()
plt.grid(True)
# Separate histogram for each variable on each axis to show univariate analysis while a scatter plot is ther to show bivariate analysis. 
# try to add hue to it.
```

## Hexbin Plots

It is a 2-d histograms as it includes denser or lighter hexagons which interpret higher or lower frequency of values, along with the separate histograms for univariate analysis on each axis.

```python
plt.style.use("ggplot")
sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex")
```

# 2-D KDE Plots

These plots includes a kde for each axis separately for univariate analysis and a 3-d plot in a 2-|d plane where the denser region
depicts the higher concentration of data points plotted while lighter regions depicts lower concentration of data points plotted.
You can also assume the denser region to be comming out of screen and is very near to you while lighter region to be away from you.

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="kde", fill=True,)
plt.style.use("seaborn-v0_8")
```

## Regression Plots

A Plot which includes a scatter plot, histogram with kde on separate axis and a regression line with a confidence interval.
Regression Line is the best fitted line segment according to the data which tells that most data points is around it and it can be used to predict future data points.
Confidence Interval is an area of lighter shade around regression line which tells that the line will fall in this particular region only(95 % confidence)

```python
sns.jointplot(x="total_bill", y="tip", kind="reg", data=tips)
plt.style.use("seaborn-v0_8-dark-palette")
```

# Pair Plots

These are used to plot pairwise multiple bivariate plots in dataset. It creates a matrix where each columns of dataset are present on x and y axis and at the point of meeting or crossing in the matrix, a plot between those particular column data is plotted. If on both axis same columns are there, then a histogram(univariate) is plotted and if different columns are there then a scatterplot is plotted(bivariate). It can be used to know which columns differentiate more.

```python
iris=sns.load_dataset("iris")
```

```python
iris.head()
```

```python
sns.pairplot(iris)
plt.style.use("seaborn-v0_8-darkgrid")
```

```python
sns.pairplot(iris, hue="species")
```

# Dist Plots

Distplots are used to do univatiate analysis on a continuous/numerical data.

Distplot combines 3 things in one graph:
1. Histogram
2. Kernel Density Enstimation (PDF)
3. Rugplot

```python
titanic=pd.read_csv("Titanic.csv")
```

```python
titanic.head()
```

```python
sns.distplot(titanic["Age"])
plt.style.use("fivethirtyeight")
```

```python
sns.distplot(titanic["Age"], hist=False)
# KDE gives a value of probability to find the value of particular variable in that region.
# Where the curve is high, the probability of finding data point is high and vice-versa.
```

## Rug PLot

```python
sns.distplot(titanic["Age"], hist=False, kde=False, rug=True)
# Rug plot tells that whether the frequency is high or not.
```

### Distplot Application

```python
#Survived vs Died
sns.distplot(titanic[titanic["Survived"]==1]["Age"])
sns.distplot(titanic[titanic["Survived"]==0]["Age"])
```
