```python
Pro_Lan=pd.read_csv("Programming_Language.csv")
```

```python
#We had used a conventional method of counter which counts when a particular item is updated or iterated.
with open('Programming_Language.csv') as csv_file:#An iterator
    csv_reader = csv.DictReader(csv_file) #Stores in the form of Dictionary

    language_counter = Counter()#The counter function

    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))# Data with key value pair in multiple tuples is stored.

languages=[]
popularity=[]
plt.figure(figsize=(9, 5)) # Width , Height

for item in language_counter.most_common(20):
    languages.append(item[0])
    popularity.append(item[1])

languages.reverse()
popularity.reverse()
    
plt.barh(languages,popularity)

plt.xlabel("Popularity")
plt.ylabel("Languages")
plt.title("Languages Vs Popularity")
plt.tight_layout()
```

```python
#This is done by using Pandas
Ids=Pro_Lan["Responder_id"]
Lang=Pro_Lan["LanguagesWorkedWith"]
language_counter = Counter()
for response in Lang:
    language_counter.update(response.split(';'))# Data with key value pair in multiple tuples is stored.

languages=[]
popularity=[]
plt.figure(figsize=(9, 5)) # Width , Height

for item in language_counter.most_common(15): #item is a tuple
    languages.append(item[0])
    popularity.append(item[1])

languages.reverse()
popularity.reverse()
    
plt.barh(languages,popularity)

plt.xlabel("Popularity")
plt.ylabel("Languages")
plt.title("Languages Vs Popularity")
plt.tight_layout()
```

```python
%matplotlib notebook
#plotting real time data
x_values=[]
y_values=[]
index=count()
#You are storing an iterator object with the help of count function which generates consecutive but infinite number of values and these values gets stored in the variable(here index).

def animate(i):
    x_values.append(next(index))
    y_values.append(random.randint(0,5))
    
    plt.cla()#Used to clear the previous plot
    plt.plot(x_values,y_values)

ani=FuncAnimation(plt.gcf(), animate, interval=1000, cache_frame_data=False)
plt.tight_layout()
# a real time but random data is plotted below.
    
```

```python
#Flappy Bird from chat gpt using FuncAnimation.
player_pos = 0
pipe_pos = 10
gravity = 0.1
jump_strength = 0.2
game_over = False

# Function to update game state and plot
def update_game(frame):
    global player_pos, pipe_pos, game_over
    
    if not game_over:
        # Update player position based on gravity
        player_pos += gravity

        # Check for collision with pipe
        if abs(player_pos - pipe_pos) < 1:
            game_over = True
        
        # Move pipe to the left
        pipe_pos -= 1

        # Check for game-over conditions
        if pipe_pos < -10:
            pipe_pos = 20

    # Clear previous plot
    plt.cla()

    # Plot player and pipe
    plt.scatter(1, player_pos, c='blue', marker='^', s=100)
    plt.bar(pipe_pos, 2, color='green')

    # Set plot limits
    plt.xlim(0, 20)
    plt.ylim(0, 10)

    # Display game-over message
    if game_over:
        plt.text(10, 5, 'Game Over', fontsize=20, ha='center')

# Create animation
ani = FuncAnimation(plt.gcf(), update_game, frames=range(100), interval=100)

plt.tight_layout()
plt.show()
```
