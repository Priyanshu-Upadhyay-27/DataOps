```python
#When two or more conditions needs to be true in masking then we define two variables and join them by '&' bitwise operator.
def two_condition_masking(name,team):
    mask1=data["player_of_match"]==name
    mask2=data["winner"]==team
    return data[mask1 & mask2]
two_condition_masking("AB de Villiers","Royal Challengers Bangalore")
#Here rows which contains player_of_match as AB de Villiers and RCB is the winner, those are printed only.
```

```python
def get_maximum_runs_scored_by_Player(batsman_name):
    m1=data2["batsman"]==batsman_name
    new_d=data2[m1]
    return new_d.groupby("bowling_team")["batsman_runs"].sum().sort_values(ascending=False)
#This snippet gives you the maximum number of runs scored against teams by a particular player.
get_maximum_runs_scored_by_Player("V Kohli")
```

```python
# isin() function
#Used when large amount of masking conditions are there.
#Below example computes the most strike rate of batsman in death_overs who had played more than 200 balls.
death_overs=data2[data2["over"]>15]
```

```python
all_batsman=death_overs.groupby("batsman")["batsman_runs"].count()
#Above statement returns a series which tells the number of balls played by each player in death_overs.
x=all_batsman>200
# x contains a boolean series which is masked below to all_batsman series
batsman_list=all_batsman[x].index.tolist()
batsman_list
```

```python
#Above list contains 43 batsman.
final=death_overs[death_overs["batsman"].isin(batsman_list)]
# We will use death_overs instead of data2 because death_overs only contains data after 16 overs.
#isin() returns a boolean series in which only those values(of death_overs) are true which are present in batsman_list.
final
```

```python
#runs
runs=final.groupby("batsman")["batsman_runs"].sum()
runs
```

```python
#balls
balls=final.groupby("batsman")["batsman_runs"].count()
balls
```

```python
sr=(runs/balls)*100
sr.sort_values(ascending=False)
```
